Cleanig data using NLP

```python
def clean_data(data):
  data = data.lower()
  data = re.sub(r'[^(a-zA-Z)\s]','', data)                
  data = re.sub(r'http\S+', " ", data)                       #to remove url
  data = re.sub(r'@\w+',' ',data)                        # to remove @
  data = re.sub(r'#\w+', ' ',data)                #to remove hastags
  data = re.sub(r'\d+()', ' ', data)            #to remove digits
  data = re.sub('r<.*?>',' ', data)               #to remove stop words
  data = data.split()
  return data
  df['tweet_text']=df['tweet_text'].apply(lambda x:clean_data(x))
```

```python
def remove_punctuation(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text
    df['tweet_text']= df['tweet_text'].apply(lambda x: remove_punctuation(x))
```

```python
def tokenization(text):
    text = re.split('\W+', text)
    return text
    df['tweet_text'] = df['tweet_text'].apply(lambda x: tokenization(x.lower()))
```

```python
import re
def clean_text(Text):
    Text = re.sub('http\S+\s*', ' ', Text)  
    Text = re.sub('RT|cc', ' ', Text) 
    Text = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', Text)  # remove punctuations in the text    
    Text = re.sub('@\S+', '  ', Text)  # remove mentions in the text
    Text = re.sub(r'[^\x00-\x7f]',r' ',Text)    
    Text = Text.lower() #changing to lower case
    Text = re.sub('\s+', ' ', Text) 
    Text = re.sub('#\S+', '', Text)  
    return Text
    
sentiment['tweet_text']=sentiment['tweet_text'].apply(clean_text)
```

Sentiment Analysis

```python
from textblob import TextBlob
```

```python
def getsubjectivity(text):
    subj=TextBlob(text).sentiment.subjectivity
    return subj

def getpolarity(text):
    return TextBlob(text).sentiment.polarity

sentiment['subjectivity']=sentiment['tweet_text'].apply(getsubjectivity)
sentiment['getpolarity']=sentiment['tweet_text'].apply(getpolarity)
```

```python
sentiment.head(2)
```

```python
def geranalysis(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
    
sentiment['Analysis']=sentiment['getpolarity'].apply( geranalysis)
```
