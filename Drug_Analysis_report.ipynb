```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import linear_model
import itertools as it
from sklearn.metrics import accuracy_score, plot_roc_curve, roc_curve, roc_auc_score, confusion_matrix, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from matplotlib.colors import ListedColormap
```

```python
df1=pd.read_parquet("train.parquet")
df2=pd.read_parquet("test.parquet")
```

```python
df1.head(5)
```

```python
df1.isnull().sum
```

```python
df1=df1.drop('Date',axis=1)
```

```python
df1.head(2)
```

```python
df1 = df1.drop_duplicates()
```

```python
train_data=pd.get_dummies(df1['Incident'], drop_first=True)
df_2=pd.concat([df1,train_data],axis=1)
```

```python
df3=df_2.drop('Incident',axis=1)
df3.head(5)
```

```python
df3['TARGET DRUG'].value_counts()
```

```python
zero=df3[df3['TARGET DRUG']==0]
one=df3[df3['TARGET DRUG']==1]
```

```python
sampe_train=zero.sample(n=9374)
sampe_traingrp=sampe_train.groupby(by='Patient-Uid',as_index=False).mean()
sampe_traingrp.head(5)
```

```python
one_datagrp=one.groupby(by='Patient-Uid',as_index=False).mean()
one_datagrp.head(2)
```

```python
traingrp1=one_datagrp.drop('Patient-Uid',axis=1)
traingrp2=traingrp1.apply(lambda x:x.astype(float))
new=pd.concat([pd.DataFrame(one_datagrp['Patient-Uid']),traingrp2],axis=1)
new.head(5)

```

```python
new_train_data=pd.concat([sampe_traingrp,new],axis=0)
```

```python
x_train=new_train_data.drop(['TARGET DRUG','DRUG_TYPE_18'],axis=1)
x_train1=x_train.drop('Patient-Uid',axis=1)
y_train=new_train_data['TARGET DRUG']
```

```python
df2.head(5)
```

```python
df2=df2.drop_duplicates()
```

```python
df21=df2.drop('Date',axis=1)
df21.head(5)
```

```python
#  OneHotEncoding:
test_data=pd.get_dummies(df2['Incident'],drop_first=True)
df21=pd.concat([df2,test_data],axis=1)
df23=df21.drop('Incident',axis=1)
df23.head(5)
```

```python
df24=df23.groupby('Patient-Uid',as_index=False).mean()
df24.tail(4)
```

```python
x_Test_data=df24.drop('Patient-Uid',axis=1)
x_Test_data.head(5)
```

```python
lr=LogisticRegression()
model_train=lr.fit(x_train1,y_train)
model_train
```

```python
predict_train=lr.predict(x_train1)
predict_train
```

```python
score=accuracy_score(predict_train,x_Test_data)
score
```
